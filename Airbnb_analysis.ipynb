```python
def remove_outliers(col_name):
    Q1 = col_name.quantile(0.25)
    Q3 = col_name.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return col_name[(col_name >= lower_bound) & (col_name <= upper_bound)]
```

```python
df_cleaned = df.copy()
df_cleaned['price'] = remove_outliers(df_cleaned['price'])
display(df_cleaned.head())
```

```python
# Group by Construction year and calculate the mean price
average_price_by_year = df.groupby('Construction year')['price'].mean().reset_index()

# Sort by Construction year for better visualization
average_price_by_year = average_price_by_year.sort_values('Construction year')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Construction year', y='price', data=average_price_by_year)
plt.title('Average Price by Construction Year')
plt.xlabel('Construction Year')
plt.ylabel('Average Price')
plt.xticks(rotation=90) # Rotate x-axis labels for better readability if there are many years
plt.tight_layout()
plt.show()
```

```python
# Show histogram of calculated host listings count distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='calculated host listings count', bins=30, kde=True)
plt.title('Distribution of Calculated Host Listings Count')
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Show histogram of price distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='price', bins=30, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Show box plot of price distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='price')
plt.title('Box Plot of Price Distribution')
plt.ylabel('Price')
plt.show()
```
